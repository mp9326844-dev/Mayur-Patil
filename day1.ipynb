```python
item1 = 1200
item2 = 800
item3 = 600
discount = 0.2
gst = 0.8

total = item1 + item2 + item3
discount_amount = total * discount
bill_after_discount = total - discount_amount

gst_amount = bill_after_discount * gst
final_bill = bill_after_discount + gst_amount

print("Total Bill:", total)
print("Final Bill after discount:", final_bill)
print("GST Amount (18%):", gst_amount)
print("Final Bill with GST:", final_bill)
```

```python
# Function to issue books
def issue_books(student_name, books):
  print(f"\n {student_name} borrowed:")
  for book in books:
    print("_", book)

# Function to calculate fine
def calculate_fine(days_late):
  fine_per_day = 5
  return days_late * fine_per_day

# Function to generate report
def student_report(student_name, books, days_late):
  issue_books(student_name, books)
  fine = calculate_fine(days_late)
  print(f"Fine to be paid: ₹{fine}\n")

# Example usage
student_report("Rahul", ["Python Basisc", "Data Science 101"], 3)
student_report("Priya", ["Machine Learning"],0)
```
