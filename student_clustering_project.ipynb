```python
#step 1:import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
data = pd.DataFrame({
    "hours_studies":np.random.randint(1, 10, 50),
    "attendence":np.random.randint(50, 101, 50),
})
data.head()
```

```python
x = data[["hours_studies","attendence"]]
model = KMeans(n_clusters=3, random_state=42, n_init=10)
data["cluster"] = model.fit_predict(x)
data.head()
```

```python
score = silhouette_score(x, data["cluster"])
print("silhouette score:",score)
```

```python
plt.scatter(data["hours_studies"],data["attendence"],c=data["cluster"])
centers= model.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],marker='x',s=200)
plt.xlabel("hours_studies")
plt.ylabel("attendence")
plt.title("student segmentation")
plt.show()
```

```python
#example:new student
# hours _studied = 6,attendence =80
new_student =([6,80])
predicted_cluster = model.predict([new_student])
print("new student belong to cluster:",predicted_cluster[0])
```
