```python
# Creating sets
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
# Add elements
set1.add(10)
# Remove elements
set1.discard(2)
# Set operations
print("Union:", set1 | set2)
print("Intersection:", set1 & set2)
print("Difference:", set1 - set2)
   # {1, 3, 4, 5, 6, 10}
   # {3, 4}
   # {1, 10}
print("Is Subset?", {3, 4}.issubset(set1)) # True
```

```python
cricket = {"Alice", "Bob", "Charlie", "David"}
football = {"Charlie", "Eva", "Frank", "Bob"}
# 1. Both sports
both = cricket & football
# 2. Only cricket
only_cricket = cricket - football
# 3. Either cricket or football
either = cricket | football
# 4. Add George
football.add("George")
# 5. Subset check
subset_check = cricket.issubset(football)
print("Both:", both)
print("Only Cricket:", only_cricket)
print("Either:", either)
print("Football Players (updated):", football)
print("Cricket ⊆ Football:", subset_check)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
# Sample dataset
df = pd.DataFrame({
'Name': ['Alice', 'Bob', 'Charlie'],
'Age': [25, 32, 47],
'Country': ['US', 'India', 'US'],
'Spend1': [200, 150, 300],
'Spend2': [100, 200, 400]
})
# Feature Engineering: Total Spend
df['Total_Spend'] = df['Spend1'] + df['Spend2']
# Feature Engineering: Age Group
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 30, 50], labels=['Young', 'Middle'])
# Encode categorical feature
encoder = LabelEncoder()
df['Country_encoded'] = encoder.fit_transform(df['Country'])

print(df)
```

```python
# Step 1: Import & EDA
import pandas as pd
from google.colab import files

# Upload the students.csv file
# If you already uploaded it, you can comment out the line below
uploaded = files.upload()
for fn in uploaded.keys():
  print(f'User uploaded file "{fn}"')

# Load dataset
df = pd.read_csv("students.csv")

# Quick look
print(df.head())
print(df.info())
print(df.describe())
# Check missing values
print(df.isnull().sum())
```

```python
# 1. Fill with Mean (Numerical)
df['Math_Score'] = df['Math_Score'].fillna(df['Math_Score'].mean())
# 2. Fill with Median
df['Writing_Score'] = df['Writing_Score'].fillna(df['Writing_Score'].median())
# 3. Fill with Mode (Categorical)
df['Parental_Education'] = df['Parental_Education'].fillna(df['Parental_Education'].mode()[0])
# 4. Forward Fill / Backward Fill
df['Reading_Score'] = df['Reading_Score'].fillna(method='ffill')
# 5. Drop rows (if too many missing)
df = df.dropna(subset=['Attendance'])
```

```python
# Total Score
df['Total_Score'] = df[['Math_Score','Reading_Score','Writing_Score']].sum(axis=1)
# Average Score
df['Average_Score'] = df['Total_Score'] / 3
# Performance Category
df['Performance_Level'] = pd.cut(
df['Average_Score'],
bins=[0,60,75,100],
labels=['Low','Medium','High']
)
# Attendance Category
df['Attendance_Level'] = pd.cut(
df['Attendance'],
bins=[0,75,90,100],
labels=['Poor','Average','Good']

)
```

```python
# Display available columns for debugging
print("DataFrame columns:", df.columns)

# Average performance by gender
print(df.groupby('Gender')['Average_Score'].mean())
# Distribution of performance levels
print(df['Performance_Level'].value_counts())
# Correlation between attendance & average score
print(df[['Attendance_Numeric','Average_Score']].corr())
# Parental education impact
print(df.groupby('Parental_Education')['Average_Score'].mean())
```
