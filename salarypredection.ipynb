```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
np.random.seed(42)
n=200
data = pd.DataFrame({
    "experience":np.random.randint(0,15,n),
    "education_year": np.random.randint(10,20,n),
    "age": np.random.randint(22,60,n),

})
data["salary"]=(
    3000+
    data["experience"]*4000+
    data["education_year"]*200+
    data["age"]* 500+
    np.random.randn(n)*10000
)
data.head()
```

```python
print(data.shape)
print(data.info)
print(data.describe())
```

```python
corr_matrix = data.corr()
plt.figure(figsize=(8, 6))
img = plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar(img)
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=45)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.title('Correlation Matrix')
plt.show()
```

```python
features = ["experience","education_year","age"]
for col in features:
  plt.figure()
  plt.scatter(data[col],data["salary"])
  plt.xlabel(col)
  plt.ylabel("salary")
  plt.show()
```

```python
x = data[["experience","education_year","age"]]
y = data["salary"]
```

```python
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state = 42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
intercept = model.intercept_
coefficients = model.coef_

print(f"Salary = {intercept:.2f} + {coefficients[0]:.2f} * experience + {coefficients[1]:.2f} * education_year + {coefficients[2]:.2f} * age")
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(mse, rmse, r2)
```

```python
new_data = pd.DataFrame({
    "experience":[5],
    "education_year":[16],
    "age":[30]
})
print(model.predict(new_data))
```
